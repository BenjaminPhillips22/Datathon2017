```python
import numpy as np
import pandas as pd
import os
import time
import matplotlib.pyplot as plt
%matplotlib inline

import some_functions
```

```python
f2 = "C:/Users/ben/Desktop/MelbDatathon2017/MelbDatathon2017/Lookups/Drug_LookUp.txt"
f3 = "C:/Users/ben/Desktop/MelbDatathon2017/MelbDatathon2017/Lookups/ChronicIllness_LookUp.txt"
f4 = "C:/Users/ben/Desktop/MelbDatathon2017/MelbDatathon2017/Lookups/patients.txt"
f5 = "C:/Users/ben/Desktop/MelbDatathon2017/fourthAttempt/remoteness2.csv"


Drug_LookUp = pd.read_table(filepath_or_buffer=f2, header=0)
ChronicIllness_LookUp = pd.read_table(filepath_or_buffer=f3, header=0)
Patients_LookUp = pd.read_table(filepath_or_buffer=f4, header=0)
remoteness = pd.read_table(filepath_or_buffer=f5, header=0, delimiter=',')
```

```python
loc1 = "C:/Users/ben/Desktop/MelbDatathon2017/MelbDatathon2017/Transactions/patients_"
loc2 = "C:/Users/ben/Desktop/MelbDatathon2017/MelbDatathon2017/MISSING_TRANSACTIONS/missing_patients_"

aList = list()

start = time.time()

for i in range(1,26):
    
    
    f1 = loc1 + str(i) + '.txt'
    df = pd.read_table(filepath_or_buffer=f1, sep='\t', header=0)
    
    f2 = loc2 + str(i) + '.txt'
    df_missing = pd.read_table(filepath_or_buffer=f2, sep='\t', header=0)
    
    
    aList.append(some_functions.extract_features(df=df, df_missing=df_missing, ChronicIllness_LookUp=ChronicIllness_LookUp, Drug_LookUp=Drug_LookUp, Patients_LookUp=Patients_LookUp, remoteness=remoteness))

    elapsed = (time.time() - start)
    print("Time: " + str(round(elapsed, 4)) + " seconds, loop: " + str(i) )
    
    
    
result = pd.concat(aList)
result.shape
```
